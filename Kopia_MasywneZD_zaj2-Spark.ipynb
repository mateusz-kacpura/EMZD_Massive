```python
lines = sc.textFile('pg27761.txt')
```

```python
def wordcount_mapper(line):
  return [(w, 1) for w in line.split()]
```

```python
def wordcount_reducer(data):
  key, values = data
  return (key, sum(values))
```

```python
result = lines.flatMap(wordcount_mapper).groupByKey().map(wordcount_reducer)
```

```python
result.map(lambda x: (x[1], x[0])).sortByKey(ascending=False).collect()
```

```python
result = lines.flatMap(wordcount_mapper).groupByKey().map(wordcount_reducer).map(lambda x: (x[1], x[0])).sortByKey(ascending=False).collect()
result
```
